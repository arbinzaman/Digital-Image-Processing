```python
# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create second image (Magenta Circle on Black Background)
img2 = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.circle(img2, (200, 200), 50, (255, 0, 255), -1)  # Magenta circle
```

```python
# Create first image (Green Rectangle on Black Background)
img1 = np.zeros((400, 400, 3), dtype=np.uint8)
cv2.rectangle(img1, (150, 150), (250, 250), (0, 255, 0), -1)  # Green rectangle
```

```python

# Perform bitwise operations
bitwise_and = cv2.bitwise_and(img1, img2)
bitwise_or = cv2.bitwise_or(img1, img2)
bitwise_xor = cv2.bitwise_xor(img1, img2)
bitwise_not1 = cv2.bitwise_not(img1)
bitwise_not2 = cv2.bitwise_not(img2)
multiply = cv2.multiply(img1, img2)
alpha_blend = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)
subtract = cv2.subtract(img1, img2)
```

```python
# Display the results
titles = [
    "Image 1 (Green Rectangle)", "Image 2 (Magenta Circle)", "Bitwise AND",
    "Bitwise OR", "Bitwise XOR", "Bitwise NOT (Image 1)", "Bitwise NOT (Image 2)",
    "Element-wise Multiply", "Alpha Blending", "Image Subtraction"
]
images = [img1, img2, bitwise_and, bitwise_or, bitwise_xor,
          bitwise_not1, bitwise_not2, multiply, alpha_blend, subtract]

plt.figure(figsize=(12, 10))
for i in range(10):
    plt.subplot(4, 3, i + 1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
